```python
#dataset

import pandas as pd
import numpy as np
data = {
    'Product': ['Laptop', 'Mobile', 'Tablet', 'Headphones', 'Smartwatch', 'Camera'],
    'Price': [55000, 20000, 30000, 2000, 8000, 45000],
    'Units_Sold': [25, 60, 35, 80, 50, 15],
    'Discount':[10,15,12,20,18,8]
}
df=pd.DataFrame(data)
print(df)
```

```python
#min max normalization
#min max normalization for price
df['Price_MinMax']=(df['Price']-df['Price'].min())/(df['Price'].max()-df['Price'].min())
print(df[['Product','Price','Price_MinMax']])
#all columns can be done together
```

```python
#normalizing multiple columns
cols=['Price','Units_Sold','Discount']
df_norm=(df[cols] - df[cols].min())/(df[cols].max()-df[cols].min())
print(df_norm)
#normalization coverts the value between 1 to 0
```

```python
#z sore optimization
#handles outliers x-xmin/xs+d
#deals with outliers
df['Unit_ZScore']=(df['Units_Sold']-df['Units_Sold'].mean())/df['Units_Sold'].std()
print(df[['Product','Units_Sold','Unit_ZScore']])
```

```python
#now normalization for 3 columns
cols=['Price','Units_Sold','Discount']
df_norm=(df[cols] - df[cols].mean())/(df[cols].std())
print(df_norm)
```

```python
#Decimal Scaling
#table- convert into decimal scaling, pick up highst value divie by 10 ntil not less thn 1
df['Price_Decimal']=df['Price']/10000
print(df[['Product','Price','Price_Decimal']])
#cant be dont on thre columns caus highet values will be diff
```

```python
df['Unit_Decimal']=df['Units_Sold']/100
print(df[['Product','Units_Sold','Unit_Decimal']])
```

```python
import pandas as pd
import numpy as np
data = {
'Order_ID':[101,102,103,104,105,106],
'Customer_Gender':['Male','Female','Female','Male','Female','Male'],
'Payment_Method':['UPI','Credit Card','Debit Card','COD','UPI','Debit Card'],
'Product_Category':['Electronics','Clothing','Home','Electronics','Beauty','Electronics'],
'City':['Pune','Mumbai','Delhi','Bangalore','Pune','Hyderabad'],
'Order_Value':[4500,2300,3200,5200,1800,4100]
}
df = pd.DataFrame(data)
display(df)
```

```python
#Method 1 :Label Encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Gender_Label']=le.fit_transform(df['Customer_Gender'])
df['City_Label']=le.fit_transform(df['City'])#gives values alphabetically

display(df)
#reverse encoding can be done too here male= 1 female=0
#label encdig can be done
```

```python
#1 Method 2: One-Hot Encoding
df_encoded=pd.get_dummies(df,columns=['Payment_Method'])
display(df_encoded)
#one hot encoder
'''     cod cc dc upi
   cod 1    0   0  0
   cc  0    1   0  0
   deb  0   0   1  0
   upi 0    0   0  1'''
```

```python
#encodig multiple columns
df_multi=pd.get_dummies(df,columns=['Customer_Gender','City'])
display(df_multi)
#gender
'''   f   m
    f  1  0
    m  0  1

    but in label encoding female is 0'''
```

```python
#dummy encoding
#instead of n+1 bit it will be n bit
#dummy encoding= label encoding for n=1
```

```python
#Method 3: Dummy Encoding(Drop First)
df_dummy=pd.get_dummies(df,columns=['Payment_Method'],drop_first=True)
display(df_dummy)
```
